We would like to have variables that contain lats, lons and wind speed for different tracks. Good way to do this is to use dictionarys.

```python
d = {'1':'hi','2':'hcu','name':'nikolay'}
```

```python
d
```

```python
d['1']
```

```python
d['name']
```

```python
d = {'track1':[1,2,3,4,5]}
```

```python
d['track1']
```

Add key-value pair to the dict:

```python
d['newkey'] = [5,6,7,8,9]
```

```python
d['newkey']
```

We can start from empty dict:


```python
d2 = {}
```

```python
d2
```

```python
d2['one'] = [2,3,4]
```

```python
d2
```

We can also append to the list in the dictionary:

```python
d2['one'].append(5)
```

```python
d2
```

We can start from empty list:

```python
d3 = {}
```

```python
d3['two'] = []
```

```python
d3
```

```python
d3['two'].append(10)
```

```python
d3
```
